## 질문1. 검색어로 찾은 책 목록을 json 파일로 저장하기


```python
import pandas as pd
import json
import os
import requests
from bs4 import BeautifulSoup
from pprint import pprint

def search_books(query):
    payload = {
        'query': query,
        'display': 50,
        'sort': "sim"
    }

    headers = {
        'X-Naver-Client-Id' : 'w7yPFLn3USNpk4tFMKiw',
        'X-Naver-Client-Secret' : '5AhLzv8ReZ'
    }



    url = f"https://openapi.naver.com/v1/search/book.json"

    res = requests.get(url, params=payload, headers=headers)
    books = res.json()['items']
    book_list = [];
    for book in books:
        if query in book['title']:
            book_list.append(book)

    if res.ok:
        folder_path = "data"
        file_path = os.path.join(folder_path, "books.json")
        os.makedirs(folder_path, exist_ok=True)

        with open(file_path, 'w', encoding='EUC-KR') as file:
            json.dump(book_list, file, ensure_ascii=False, indent=2)
    else:
        print(f'Error code = {res.status_code}')

```
